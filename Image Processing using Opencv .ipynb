```python
cv2.circle(img=bimg,
           center=(4000,1000),
           radius=500,
           color=(0,255,0),
           thickness=-1)
plt.imshow(bimg)
```

### Draw a Rectangle

```python
cv2.rectangle(bimg,
              pt1=(200,200),
              pt2=(300,300),
              color=(255,255,255),
              thickness=5)
plt.imshow(bimg)
```

### Draw Tri

```python
ver=np.array([[200,100],[100,100],[100,200]])
pts=ver.reshape(-1,1,2)
```

```python
cv2.polylines(bimg,
              [pts],
              isClosed=True,
              color=(255,255,255),
              thickness=3)
```

```python
plt.imshow(bimg)
```

### Fill Triangle

```python
vertices=np.array([[200,100],[100,100],[100,200]])#change y to bring filling out
pts=vertices.reshape(-1,1,2)
```

```python
cv2.polylines(bimg,[pts],color=(255,255,255),isClosed=True)
```

```python
plt.imshow(bimg)
```

### Draw Line


```python
bimg=cv2.line(bimg,pt1=(200,100),pt2=(300,400),color=(255,255,255),thickness=10)
plt.imshow(bimg)
```

## Drawing_with_mouse

```python
import cv2
import numpy as np

# x, y, flag, pr are feed from OpenCV automatically
def draw_circle(event,x,y,flag,pr):
    if event==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),70,(35,69,78),-1)
    
#connect the func with callback
cv2.namedWindow(winname='mickey')

#callback
cv2.setMouseCallback('mickey',draw_circle)

#use openCV to show image
img=np.zeros((512,512,3),np.int8)

while True:
    cv2.imshow('mickey',img)
    if cv2.waitKey(5) & 0xFF==ord('q'):
        break
        
cv2.destroyAllWindows()
```

## Add choice to mouse

```python
import cv2
import numpy as np

# x, y, flag, pr are feed from OpenCV automatically
def draw_circle(event,x,y,flag,pr):
    #Left button down
    if event==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(75,131,251),-1)
    # Right button down
    elif event==cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),-1)
        
#connect the func with callback
cv2.namedWindow(winname='mickey')

#callback
cv2.setMouseCallback('mickey',draw_circle)

#use openCV to show img
img=np.zeros((512,512,3),np.int8)

while True:
    cv2.imshow('mickey',img)
    if cv2.waitKey(5) & 0xFF==ord('q'):
        break
        
cv2.destroyAllWindows()
```

# Mouse Functions

```python
'''
True when mouse button is down
False when mouse button is up
'''
drawing=False
ex=-1
ey=-1

# x, y, flag, pr are feed from OpenCV automatically
def draw_rectangle(event,x,y,flag,pr):
    global ex, ey, drawing
    
    if  event==cv2.EVENT_LBUTTONDOWN:
        drawing=True
        ex,ey=x,y
        
    elif event==cv2.EVENT_MOUSEMOVE:
        if drawing==True:
            cv2.rectangle(img,(ex,ey),(x,y),(255,0,255),-1)
        
    elif event==cv2.EVENT_LBUTTONUP:
        drawing=False
        cv2.rectangle(img,(ex,ey),(x,y),(255,0,255),-1)
             
  
#Connect function with a callback
img=np.zeros((512,512,3),np.int8)

cv2.namedWindow(winname='my_drawing')

#Callback
cv2.setMouseCallback('my_drawing',draw_rectangle)

#Using OpenCV to show image
while True:
    cv2.imshow('my_drawing',img)
    
    if cv2.waitKey(5) & 0xFF==ord('q'):
        break
        
        
cv2.destroyAllWindows()       
```
